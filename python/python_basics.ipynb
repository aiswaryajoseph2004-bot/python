```python
#write a python program to concatenate following dictionaries to create a new one
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 600, 'b': 200, 'd': 500}
d3 = {}

# Merge values for common keys and retain unique keys
for i, j in d1.items():
    if i in d2:
        d3[i] = j + d2[i]
    else:
        d3[i] = j

for x, y in d2.items():
    if x not in d1:
        d3[x] = y

print(d3)
```

```python
class person:
  def __init__(self,name):
    self.name=name

class Employee(person):
  def __init__(self,name,salary):
    super().__init__(name)
    self.salary=salary
class job:
  def __init__(self,salary):
    self.salary=salary

class Employeepersonjob(Employee,job):
  def __init__(self, name, salary):
    Employee.__init__(self,name,salary)
    job.__init__(self,salary)

class manager(Employeepersonjob):
  def __init__(self, name, salary,department):
    Employeepersonjob.__init__(self,name,salary)
    self.department=department


class Assistantmanager(Employeepersonjob):
    def __init__(self, name, salary, team_size):
        Employeepersonjob.__init__(self, name, salary)
        self.team_size = team_size

class Seniormanager(manager,Assistantmanager):
    def __init__(self, name, salary, department, team_size):
        manager.__init__(self, name, salary, department)
        Assistantmanager.__init__(self, name, salary, team_size)



emp = Employee("John", 40000)
print(emp.name, emp.salary)


emp2 = Employeepersonjob("Alice", 50000)
print(emp2.name, emp2.salary)

mgr = manager("Bob", 60000, "HR")
print(mgr.name, mgr.salary, mgr.department)

asst_mgr = Assistantmanager("Charlie", 45000, 10)
print(asst_mgr.name, asst_mgr.salary, asst_mgr.team_size)

sen_mgr = Seniormanager("David", 70000, "Finance", 20)
print(sen_mgr.name, sen_mgr.salary, sen_mgr.department, sen_mgr.team_size)
```

```python
from abc import ABC,abstractmethod

class shape(ABC):
  @abstractmethod
  def calculate_area(self):
    pass

class circle(shape):
  def __init__(self,radius):
    self.radius=radius

  def calculate_area(self):
    print(f"circle area is :{3.14 * (self.radius)**2}")

class Rectangle(shape):
  def __init__(self,width,height):
    self.width=width
    self.height=height

  def calculate_area(self):
    print(f"Rectangle area is : {self.width * self.height}")

cir=circle(10)
rect=Rectangle(5,10)

cir.calculate_area()
print()
rect.calculate_area()
```

**Write a program that opens a file (if it exists), reads content, and closes it in the finally block. Ensure the file is always closed even if an error occurs.**

```python
try:
    filename = input("Enter the filename: ")
    file = open(filename, "r")
    content = file.read()
    print("\nFile Content:\n")
    print(content)

except FileNotFoundError:
    print("Error: The file does not exist.")
except IOError:
    print("Error: Could not read the file.")
finally:
    try:
        file.close()
        print("\nFile closed successfully.")
    except NameError:
       print("NameError Occurred and Handled")

```

## Two Threading


```python
import threading
import time
# function for the chef's work (Thread 1)
def chef_work():
  for i in range(5):            # Chef prepares 5 dishes
    print(f"chef:preparing dish {i+1}...")
    time.sleep(2)             # Simulate time taken to prepare a dish
    if i==2:        # Assistant starts washing utensils during the 3rd dish
      print("chef:calling assistant to wash utensils.")
      time.sleep(1)           # Simulate time taken to call the assistant
      assistant_thread=threading.Thread(target=assistant_work)
      assistant_thread.start()
      assistant_thread.join()     # Wait for the assistant to finish
      print("chef:Assistant done.resuming dish preparation.")
  print("chef:All dishes are prepared")


# Function for the assistant's work (Thread 2)
def assistant_work():
  for j in range(3):      # Assistant washes 3 utensils
    print(f"Assistant: washing utensil {j+1}...")
    time.sleep(2)           # Simulate time taken to wash each utensil
    print("Assistant: All utensils are clean.")

print("kitchen:starting work.")
chef_thread=threading.Thread(target=chef_work)
chef_thread.start()
chef_thread.join()      # Wait for the chef to complete their work
print("kitchen:All work completed.")
```

Simulate video streaming using threads:

Video plays for 8 seconds (printing Playing frame X).

At the 5th second, start a buffering thread (takes 4 seconds).

Video playback must pause until buffering completes, then resume.

```python
import threading
import time

def video_player():
  print(f"video player:starting the video playback")
  for i in range(8):
    print(f"video player:playing frame {i+1}")
    time.sleep(1)
    if i==4:
      print(f"video player:buffering start")
      time.sleep(1)
      buff_thread=threading.Thread(target=buffering)
      buff_thread.start()
      buff_thread.join()
      print(f"video player:buffering complete,resuming playback.")
def buffering():
    for j in range(4):
        print(f"Buffering... {j+1} seconds")
        time.sleep(1)
video_thread = threading.Thread(target=video_player)
video_thread.start()
video_thread.join()
print("Video Streaming Completed.")
```

Train Journey & Ticket Checking
Write a program to simulate a train journey:

The train runs for 10 seconds (print Train running second X).

At the 6th second, start a thread for ticket checking (3 seconds).

The train journey must wait until ticket checking is done, then continue.

```python
import threading
import time


def train_journey():
  print(f"train_journey:starting train journey")
  for i in range(10):
    print(f" train_journey:train running second {i+1}")
    time.sleep(1)
    if i==3:
      print(f"train_journey:ticket checking start")
      time.sleep(1)
      tic_thread=threading.Thread(target=ticket_checking)
      tic_thread.start()
      tic_thread.join()
      print(f"train_journey:ticket checking is done,continue")
def ticket_checking():
  for j in range(3):
        print(f"checking... {j+1} seconds")
        time.sleep(1)
train_thread = threading.Thread(target=train_journey)
train_thread.start()
train_thread.join()
print("train journey is completed...")
```
