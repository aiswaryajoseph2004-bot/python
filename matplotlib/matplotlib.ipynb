## set font properties for title and lables

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':10}
font2={'family':'serif','color':'darkred','size':10.5}
plt.title("sports watch data",fontdict=font1)
plt.xlabel("average pulse",fontdict=font2)
plt.ylabel("calorie burnage",fontdict=font2)
plt.plot(x,y)
plt.show()
```

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
fig = plt.figure(
    FigureClass = Waffle,
    rows=5,
    columns = 10,
    values = [48,46,6],
    figsize = (5,3)
)
plt.show()
```

```python
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=10,
    values=[48,46,6],
    figsize=(5,3)
)
plt.show()
```

## values in dict & auto sizing

```python
from matplotlib import legend
data={'Democratic':48,'Republican':46,'Libertarian':6}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    legend={'loc':'upper left','bbox_to_anchor':(1.1,1)}
)
plt.show()
```

## Title, Legend, Colors, Background Color, Block Color, Direction and Style


```python
data={'Democratic':48,'Republican':46,'Libertarian':6}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    colors=["red","blue","black"],
    title={"label":'vote Percentage in 2016 US presidential Election','loc':'left'},
    labels=[f'{k}({v}%)' for k,v in data.items()],
    legend={'loc':'lower left','bbox_to_anchor':(0,-0.4),'ncol':len(data),'framealpha':0},
    starting_location='NW',
    block_arranging_style='snake')
fig.set_facecolor('#EEEEEE')
plt.show()
```

### plot with icons pictogram chart

```python
data={'Democratic':48,'Republican':46,'Libertarian':6}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    colors=["red","blue","black"],
    legend={'loc':'upper left','bbox_to_anchor':(1,1)},
    icons='child',
    font_size=12,
    icon_legend=True)
plt.show()
```
